```python
import pandas as pd
```

```python
data = pd.read_csv("D:/AI - Exercise/VietAI-x-CoderSchool/Week5/BTVN_3/dataset_3.csv")
```

```python
#  Task 1: Liệt kê các sản phẩm có doanh thu lớn hơn 100.000 Euro
data["revenue"] = data["price"] * data["units_sold"]
data[data["revenue"] > 100000]
```

```python
#  Task 2: Liệt kê các sản phẩm có units_sold nằm trong top 100, đồng thời nằm trong nhóm 10% sản phẩm có rating cao nhất

top100 = data.sort_values(by="units_sold", ascending=False).head(100)
top_rating = data.sort_values(by="rating", ascending=False).head(int(len(data)*0.1))
task2 = pd.merge(top100, top_rating, on="title", how="inner")
print(task2)
```

```python
#  Task 3: Có phải sản phẩm càng được rating nhiều thì càng bán được nhiều sản phẩm ko
# => Đúng
task3 = data.groupby(by="rating")["units_sold"].sum()
print(task3)
corr = data["rating_count"].corr(data["units_sold"])
print(f'Hệ số tương quan là: {corr}')
```

```python
#  Task 4: Thống kê Top 50 sản phẩm nằm trong rating từ 3.0 → 4.0 bán chạy nhất
data_task4 = data[(data["rating"] <= 4) & (data["rating"] >= 3)]
task4 = data_task4.sort_values(by="units_sold", ascending=False).head(50)
print(task4)
```

```python
#  Task 5: Có bao nhiêu sản phẩm có giá bán lớn hơn 50 EURO
# price lớn nhất là 49.
task5 = data[data["price"] > 50]
print(len(task5))
temp = data[data["retail_price"] > 50] # dùng retail_price 
print(len(temp))
```
