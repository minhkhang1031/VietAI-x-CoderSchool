```python
import pandas as pd
```

```python
data = pd.read_csv("D:/AI - Exercise/VietAI-x-CoderSchool/Week5/BTVN_1/dataset_1.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
# Xóa dấu , trong Population, Density KM2	Density M2
def remove_comma(x):
    return x.replace(",","")

data["Population"] = data["Population"].apply(lambda x: remove_comma(x))
data["Density KM2"] = data["Density KM2"].apply(lambda x: remove_comma(x))
data["Density  M2"] = data["Density  M2"].apply(lambda x: remove_comma(x))
```

```python
# Chuyển đổi dữ liệu từ object sang các dạng phù hợp
data["Rank"] = data["Rank"].astype("int")
data["City"] = data["City"].astype("string")
data["Population"] = pd.to_numeric(data["Population"], errors='coerce')
data['Area KM2'] = pd.to_numeric(data['Area KM2'], errors='coerce')
data['Area   M2'] = pd.to_numeric(data['Area   M2'], errors='coerce')
data["Density KM2"] = pd.to_numeric(data["Density KM2"], errors='coerce')
data["Density  M2"] = pd.to_numeric(data["Density  M2"], errors='coerce')
data["Country"] = data["Country"].astype("string")
```

```python
print(data.to_string())
```

```python
# Thay đổi giá trị index 62
x = data["Area KM2"].median()
y = data["Area   M2"].median()
data.loc[62, "Area KM2"] = x
data.loc[62, "Area   M2"] = y
```

```python
#  Task 1: Sắp xếp và in ra 10 thành phố có dân số lớn nhất & 10 thành phố có dân số nhỏ nhất

city_large = data.sort_values(by="Population", ascending=False).head(10) # lớn nhất
city_small = data.sort_values(by="Population", ascending=True).head(10) # nhỏ nhất

task1 = pd.concat([city_large, city_small])
print(task1.to_string())
```

```python
#  Task 2: In ra tên các quốc gia có tối thiểu 3 thành phố trong danh sách này
task2 = data.groupby("Country").size()
print(task2[task2>=3])
```

```python
#  Task 3: In ra Top 5 quốc gia có nhiều thành phố xuất hiện trong bảng này nhất
task3 = task2.sort_values(ascending=False).head(5)
print(task3)
```

```python
# Task 4: In ra các thành phố có dân số & diện tích đều nằm trong Top 20
data.sort_values(by=["Population","Area KM2"], ascending=False).head(20)
```

```python
#  Task 5: Thống kê mật độ dân số theo quốc gia
sum_density = data.groupby(by="Country")["Density KM2"].sum().reset_index() # mật độ dân số của quốc gia
mean_density = data.groupby(by="Country")["Density KM2"].mean().reset_index() # mật độ dân số bình quân 
print(sum_density)
print(mean_density)
```

```python
#  Task 6: Thống qua các thành phố có dân số lớn nhất của từng quốc gia (chỉ tính của những quốc gia có 2
# thành phố xuất hiện trở lên trong bảng)

temp = task2[task2 >= 2].index

country_more_2_city = data[data["Country"].isin(temp)]

task6 = country_more_2_city.loc[country_more_2_city.groupby(by="Country")["Population"].idxmax()]

print(task6.to_string())
```
