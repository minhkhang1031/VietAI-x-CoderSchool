```python
# Bài 3a: Áp dụng với các bạn chưa/ít kinh nghiệm về lập trình
# Nhập input đầu vào là 2 thông số, chiều dài & chiều rộng của 1 bông hoa bất kỳ (theo đơn vị mm). Xác định xem bông hoa đấy thuộc loại hoa nào 
# (Iris setosa, Iris virginica hay Iris versicolor). Loại của bông hoa được xác định là class của
# a) Điểm có tọa độ gần nó nhất
#a input [x,y] -> tính khoảng cách giữa điểm mới và điểm trong dataset -> 1NN
data_setosa_3 = np.array(data_setosa[["sepal.length", "sepal.width"]])
data_versicolor_3 = np.array(data_versicolor[["sepal.length", "sepal.width"]])
data_virginica_3 = np.array(data_virginica[["sepal.length", "sepal.width"]])

def manhattan_distance(x,y):
    return (np.absolute(x[0] - y[0]) + np.absolute(x[1] - y[1]))
    
def coordinate(length, width):
    if length > 0 and width > 0:
        c_setosa = []
        c_versicolor = []
        c_virginica = []
        new_value = np.array([length, width])
        for i in data_setosa_3:
            c_setosa.append(manhattan_distance(new_value,i))
        for i in data_versicolor_3:
            c_versicolor.append(manhattan_distance(new_value,i))
        for i in data_virginica_3:
            c_virginica.append(manhattan_distance(new_value,i))
        min_setosa = np.min(c_setosa)
        min_versicolor = np.min(c_versicolor)
        min_virginica = np.min(c_virginica)
        if min_setosa < min_versicolor and min_setosa < min_virginica:
            return "Iris Setosa"
        if min_versicolor < min_setosa and min_versicolor < min_virginica:
            return "Iris Versicolor"
        if min_virginica < min_versicolor and min_virginica < min_setosa:
            return "Iris Virginica"
        return "Khong the tim duoc"
        
# b) Trung bình cộng nhỏ nhất đến tất cả các điểm của từng loại
def min_mean(length, width):
    if length > 0 and width > 0:
        c_setosa = []
        c_versicolor = []
        c_virginica = []
        new_value = np.array([length, width])
        for i in data_setosa_3:
            c_setosa.append(manhattan_distance(new_value,i))
        for i in data_versicolor_3:
            c_versicolor.append(manhattan_distance(new_value,i))
        for i in data_virginica_3:
            c_virginica.append(manhattan_distance(new_value,i))
        mean_setosa = np.mean(c_setosa)
        mean_versicolor = np.mean(c_versicolor)
        mean_virginica = np.mean(c_virginica)
        print("Mean Setosa:", mean_setosa)
        print("Mean Versicolor:", mean_versicolor)
        print("Mean Virginica:", mean_virginica)
        if mean_setosa < mean_versicolor and mean_setosa < mean_virginica:
            return "Iris Setosa"
        if mean_versicolor < mean_setosa and mean_versicolor < mean_virginica:
            return "Iris Versicolor"
        if mean_virginica < mean_versicolor and mean_virginica < mean_setosa:
            return "Iris Virginica"
        return 0
```

```python
#  Bài 3b: Áp dụng cho các bạn đã có kinh nghiệm về lập trình
# Nhập input đầu vào là 2 thông số, chiều dài & chiều rộng của 1 bông hoa bất kỳ (theo đơn vị mm). Xác định xem bông hoa đấy thuộc loại hoa nào 
# (Iris setosa, Iris virginica hay Iris versicolor). Loại của bông hoa được xác định là class của
# a) Điểm có tọa độ gần nó nhất. Nếu có 2 điểm thuộc 2 loại hoa khác nhau cùng nằm gần nhất với input, thì xác định sang điểm kế tiếp

def find_min(x,y,z):
    if x < y and x < z:
        return x
    if y < x and y < z:
        return y
    if z < x and z < y:
        return z
    return 0

def coordinate_2nn(length, width):
    if length > 0 and width > 0:
        d_setosa = []
        d_versicolor = []
        d_virginica = []
        rs = []
        n = 2
        new_value = np.array([length, width])
        for i in data_setosa_3:
            d_setosa.append(manhattan_distance(new_value,i))
        for i in data_versicolor_3:
            d_versicolor.append(manhattan_distance(new_value,i))
        for i in data_virginica_3:
            d_virginica.append(manhattan_distance(new_value,i))
            
        d_setosa = np.sort(d_setosa)
        d_versicolor = np.sort(d_versicolor)
        d_virginica = np.sort(d_virginica)
        
        while n > 0:
            min_setosa = d_setosa[0]
            min_versicolor = d_versicolor[0]
            min_virginica = d_virginica[0]
            temp = find_min(min_setosa, min_versicolor, min_virginica)
            if temp != 0:
                if temp == min_setosa:
                    rs.append("Setosa")
                    d_setosa.pop(0)
                elif temp == min_versicolor:
                    rs.append("Versicolor")
                    d_versicolor.pop(0)
                elif temp == min_virginica:
                    rs.append("Viginica")
                    d_virginica.pop(0)
                n -= 1
            else:
                n += 1
        values, counts = np.unique(rs, return_counts=True)
        return values[np.argmax(counts)]

```

```python
#3b
#  b) 7 điểm gần nó nhất (Vote đa số). Nếu không xác định được class chiếm đa số, in ra bông hoa này ko thuộc class nào
def coordinate_7nn(length, width):
    if length > 0 and width > 0:
        d_setosa = []
        d_versicolor = []
        d_virginica = []
        rs = []
        n = 7
        new_value = np.array([length, width])
        
        for i in data_setosa_3:
            d_setosa.append(manhattan_distance(new_value,i))
        for i in data_versicolor_3:
            d_versicolor.append(manhattan_distance(new_value,i))
        for i in data_virginica_3:
            d_virginica.append(manhattan_distance(new_value,i))
            
        d_setosa = np.sort(d_setosa)
        d_versicolor = np.sort(d_versicolor)
        d_virginica = np.sort(d_virginica)
        
        while n > 0:
            min_setosa = d_setosa[0]
            min_versicolor = d_versicolor[0]
            min_virginica = d_virginica[0]
            temp = find_min(min_setosa, min_versicolor, min_virginica)
            if temp == min_setosa:
                rs.append("Setosa")
                d_setosa.pop(0)
            elif temp == min_versicolor:
                rs.append("Versicolor")
                d_versicolor.pop(0)
            elif temp == min_virginica:
                rs.append("Viginica")
                d_virginica.pop(0)
            n -= 1
            
        values, counts = np.unique(rs, return_counts=True)
        if len(values) > 0 and len(counts) > 0 and np.max(counts) > n / 2:
            return values[np.argmax(counts)]
        else:
            return "Khong phu hop voi hoa nao"
```

```python
#3b
# c) Lấy tọa độ của input là tâm của 1 đường tròn, có bán kính 2 mm. Vote đa số dựa trên tất cả các điểm nằm trên và trong đường tròn. Nếu có 2 class có số
# điểm bằng nhau và đồng cao nhất, in ra thông báo bông hoa này ko thuộc class nào.
#c khoang cach giua diem moi xuat hien va cac diem gan do < 2 => trong ban kinh 2mm. Luu tat ca < 2mm -> 

def coordinate_radius(length, width):
    if length > 0 and width > 0:
        d_setosa = []
        d_versicolor = []
        d_virginica = []
        rs = []
        radius = 2
        new_value = np.array([length, width])
        for i in data_setosa_3:
            if manhattan_distance(new_value,i) < radius:
                d_setosa.append("Setosa")
        for i in data_versicolor_3:
            if manhattan_distance(new_value,i) < radius:
                d_versicolor.append("Versicolor")
        for i in data_virginica_3:
            if manhattan_distance(new_value,i) < radius:
                d_virginica.append("Virginica")
        rs = d_setosa + d_versicolor + d_virginica
        
        values, counts = np.unique(rs, return_counts=True)
        if len(values) > 0 and len(counts) > 0:
            c = np.max(counts)
            class_flower_max = values[counts == c]
            if len(class_flower_max) > 1:
                return "Khong phu hop voi hoa nao"
            else:
                return class_flower_max
        else:
            return "Khong phu hop voi hoa nao"
            
```
